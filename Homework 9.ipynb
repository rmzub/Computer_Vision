### Step 0

Run the necessary imports.

```python
import dlib
import cv2     
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [15, 10]
```

### Step 1

Load an image (any image that contains faces).

```python
img = cv2.imread('/Users/Roman Zubrytskyi/Downloads/2024-11-06 16.38.55.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

### Step 2

Load the dlib face predictor.

```python
detector = dlib.get_frontal_face_detector()
```

### Step 3

Run the predictor on your image.

```python
faces = detector(img, 1)
print(f"Number of faces detected: {len(faces)}")
```

### Step 4

Draw bounding boxes around the detected faces and plot the image. Use different colour for each face.

```python
image_with_detections = img.copy()

colors = [
    (255, 0, 0),    # Red
    (0, 255, 0),    # Green
    (0, 0, 255),    # Blue
    (255, 255, 0),  # Cyan
    (255, 0, 255),  # Magenta
    (0, 255, 255),  # Yellow
    (128, 0, 128),  # Purple
    (128, 128, 0),  # Olive
    (0, 128, 128),  # Teal
]

for i, face in enumerate(faces):
    x1 = face.left()
    y1 = face.top()
    x2 = face.right()
    y2 = face.bottom()
    color = colors[i % len(colors)]
    cv2.rectangle(image_with_detections, (x1, y1), (x2, y2), color, 2)

plt.figure(figsize=(12, 8))
plt.imshow(image_with_detections)
plt.axis('off')
plt.show()
```
